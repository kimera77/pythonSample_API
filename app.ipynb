```python
url = "https://swapi.info/api/planets"
respuesta = requests.get(url)
df_planetas = pd.DataFrame(respuesta.json())
```

```python
# def obtener_planeta_por_tipo(tipo):
#     lista_planetas_json = respuesta.json() 
#     for planeta in lista_planetas_json:
#         if planeta['type'].lower() == tipo.lower():
#             return planeta
#     return None

def obtener_planeta_por_tipo(tipo):
    planetas_encontrados = df_planetas[df_planetas['terrain'].str.lower().str.contains(tipo.lower())]
    return planetas_encontrados
```

```python
# First let's see what columns df_planetas has
df_planetas.head()
```

```python
obtener_planeta_por_tipo("grasslands")
```

```python
# Gradio interface
demo = gr.Interface(
    fn=obtener_planeta_por_tipo,
    inputs=gr.Textbox(label="Enter a terrain type", placeholder="example: desert, grasslands, mountains"),
    outputs=gr.Dataframe(label="Planets found"),
    title="Star Wars Planet Search",
    description="Enter a terrain type to find planets that contain it."
)
 
demo.launch()
```
